### 8.1 차원의 저주

데이터의 차원이 증가할수록 데이터 포인트 간의 거리 또한 증가하게 된다.   

차원이 높아질수록 평균거리가 점점 길어지므로 Train Data Set과 Test Data Set 간의 거리가 길어지고 이에 따라 굉장히 많은 양의 Data Set Size가 아닐 경우 Train Data Set에 맞춤형이 되버리는 Overfitting이 되어버리고 모델이 복잡해지게 됩니다. 

이러한 차원의 저주를 해결하기 위한 방법 중 하나는 데이터의 밀도가 높아질 때까지 학습 데이터셋의 크기를 늘리는 것이다.   

하지만, 데이터셋의 크기에 비해 차원은 기하급수적으로 커지기 때문에 매우 힘든일이라 할 수 있다.   

예를 들어 차원이 100개라고 하면 모든 차원에 걸쳐 균일하게 퍼져있다는 가정하에 Train Sample 간의 거리를 평균 0.1이내에 위치시키기 위해서는 관측 가능한 우주에 있는 원자 수 모두를 합친 것보다 많은 훈련 Data Sample이 있어야합니다.

### 8.3 PCA

**주성분 분석(PCA**, Principal Component Analysis**)**은 가장 대표적인 차원 축소 알고리즘이다. PCA는 먼저 데이터에 가장 가까운 초평면(hyperplane)을 구한 다음, 데이터를 이 초평면에 투영(projection)시킨다.

##### 8.3.1 분산 보존

![](./images/pca01.png)

- PCA는 projection을 하며 차원을 축소하는 개념인데 projection을 할 시에 원본 data의 손실을 최소화해야하기에 기존에 있는 분산을 최대한 유지하면서 projection해야하므로 기존 분산의 70 ~ 90%를 유지하는 것이 좋습니다. (이 책에서는 95%라고 하지만 다변량 통계수업과 데이터분석개론 수업에서는 70~90%가 적당하다고 배웠습니다.)
- PCA는 Data의 분산 유지를 최대로 하는 축을 찾는다. 즉, 원본 Data Set과 Projection된 Data Set 간의 **평균제곱거리**를 **최소화** 하는 축을 찾는다.   
 (원본 Data Set과 Projection된 Data Set 간의 평균 제곱거리 == Reconstruction Error, 재구성 오차)

##### 8.3.2 주성분

PCA는 다음과 같은 단계로 이루어진다. 

1. 학습 Data Set에서 분산 유지를 최대로 하는 축(axis)을 찾는다. 
2. 이렇게 찾은 첫번째 축과 직교(orthogonal)하면서 분산 유지를 최대로 하는두 번째 축을 찾는다.
3. 첫 번째 축과 두 번째 축에 직교하고 분산을 최대한 보존하는 세 번째 축을 찾는다.
4. `1~3`과 같은 방법으로 데이터셋의 차원(특성 수)만큼의 축을 찾는다.

=> SVD(Singular Value Decompostion, 특이값 분해)를 활용하여 PC를 찾는다.

```python
# 방법 1
pca = PCA()
pca.fit(X)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1

pca = PCA(n_components=d)
pca.fit(X)

print('principal component vec :\n', pca.components_)
```

# 배운 내용을 토대로 Coding

```python
# 방법 1
pca = PCA()
pca.fit(X)
cumsum = np.cumsum(pca.explained_variance_ratio_)
# print(cumsum,"\n\n\n")
cumsum2 = pd.DataFrame(cumsum.reshape(-1, 1), columns = ['분산누적비율'])

# 필요 차원의 갯수
d = list(cumsum2[ (cumsum2['분산누적비율'] >= 0.7) ].index)[:-1]
#  list(sss2[(sss2['분산누적비율'] >= 0.7) & (sss2['분산누적비율'] <= 0.9)].index) test용
d = list(map(lambda x: x + 1, d))

# 차원 갯수별 축

result = pd.DataFrame()

for i in range(len(d)) :
    pca = PCA(n_components= d[i])
    pca.fit(X)
    #print('차원이', d[i],'개일 때 principal component vec :\n', pca.components_)
    #print("\n\n")

    
    for j in range(len(pca.components_)) :
        
        temp_li = []
        
        temp_li.append(str(d[i]) + '개')
        temp_li.append('Y' + str(j+1))
            
        for k in range(len(pca.components_[j])) :
            temp_li.append(pca.components_[j][k])
        temp_li_df = pd.DataFrame(temp_li).T
        
        if( 'result' in globals() ) :
            result = pd.concat([result, temp_li_df], axis =0 )    
        else :
            result = temp_li_df.copy()
    
#temp_li_df
result
```

##### 8.3.8 점진적 PCA

일반적인 PCA 알고리즘은 모든 훈련 데이터 세트를 메모리에 올린 뒤 연산해야한다는 단점이 있지만 점진적 PCA - incremental PCA 의 등장으로 해소되었다.   
- 점진적 PCA(Incremental PCA) : 학습 데이터셋을 미니배치로 나눈 뒤 IPCA 알고리즘에 하나의 미니배치를 입력으로 넣어준다.

```python
from sklearn.decomposition import IncrementalPCA

n_batches = 100
inc_pca = IncrementalPCA(n_components=154)
for batch_x in np.array_split(train_x, n_batches):
    inc_pca.partial_fit(batch_x)

X_reduced = inc_pca.transform(train_x)
```

```python
rnd_pca = PCA(n_components=154, svd_solver="randomized", random_state=42)
X_reduced = rnd_pca.fit_transform(train_x)
```

### 8.4 커널 PCA

- **Kernel PCA**(KPCA) : 비선형 투영으로 차원을 축소하는 것
- 여러 가지 커널의 kPCA를 사용해 2D로 축소시킨 스위스 롤
- kernel의 종류에는 : 선형, RBF(Radial Basis Function 또는 Gaussian Kernel),Sigmoid, poly, cosine, precomputed 등

```python
from sklearn.datasets import make_swiss_roll

X, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)

from sklearn.decomposition import KernelPCA

rbf_pca = KernelPCA(n_components = 2, kernel="rbf", gamma=0.04)
X_reduced = rbf_pca.fit_transform(X)

lin_pca = KernelPCA(n_components = 2, kernel="linear", fit_inverse_transform=True)
rbf_pca = KernelPCA(n_components = 2, kernel="rbf", gamma=0.0433, fit_inverse_transform=True)
sig_pca = KernelPCA(n_components = 2, kernel="sigmoid", gamma=0.001, coef0=1, fit_inverse_transform=True)

y = t > 6.9

plt.figure(figsize=(11, 4))
for subplot, pca, title in ((131, lin_pca, "선형 커널"), (132, rbf_pca, "RBF 커널, $\gamma=0.04$"), (133, sig_pca, "시그모이드 커널, $\gamma=10^{-3}, r=1$")):
    X_reduced = pca.fit_transform(X)
    if subplot == 132:
        X_reduced_rbf = X_reduced
    
    plt.subplot(subplot)
    plt.title(title, fontsize=14)
    plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=t, cmap=plt.cm.hot)
    plt.xlabel("$z_1$", fontsize=18)
    if subplot == 131:
        plt.ylabel("$z_2$", fontsize=18, rotation=0)
    plt.grid(True)

plt.show()
```

### 8.6 다른 차원 축소 기법

- **다차원 스케일링**(MDS): MDS는 데이터 포인트 간의 거리를 보존하면서 차원을 축소하는 기법이다. 
- **Isomap**: Isomap은 각 데이터 포인트를 가장 가까운 이웃과 연결하는 식의 그래프를 만든 후 그래프에서 두 노드 사이의 최단 경로를 이루는 노드의 수인  **geodesic distance**를 유지 하면서 차원을 축소한다.
- **t-SNE**(t-distributed Stochastic Neighbor Embedding): t-SNE는 비슷한 데이터는 가까이, 비슷하지 않은 데이터는 멀리 떨어지도록 차원을 축소한다. 주로 시각화에 많이 사용되며, 특히 고차원 공간에 있는 데이터의 군집을 시각화할 때 사용한다.
- **선형 판별 분석**(LDA): LDA는  Supervised learning이며, 분류 알고리즘에 속한다. LDA는 학습 단계에서 클래스를 가장 잘 구분하는 축을 학습하며, 이 축은 데이터가 투영되는 초평면을 정의하는 데 사용할 수 있다. 이러한 초평면으로 데이터를 투영하게 되면 클래스 간의 거리를 멀리 떨어지게 축소할 수 있다.

```python
from sklearn.datasets import make_swiss_roll

X, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=41)
```

```python
from sklearn.manifold import MDS, Isomap, TSNE

# MDS
mds = MDS(n_components=2, random_state=42)
X_reduced_mds = mds.fit_transform(X)

# Isomap
isomap = Isomap(n_components=2)
X_reduced_isomap = isomap.fit_transform(X)

# t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_reduced_tsne = tsne.fit_transform(X)
```

```python
titles = ["MDS", "Isomap", "t-SNE"]

plt.figure(figsize=(11,4))

for subplot, title, X_reduced in zip((131, 132, 133), titles,
                                     (X_reduced_mds, X_reduced_isomap, X_reduced_tsne)):
    plt.subplot(subplot)
    plt.title(title, fontsize=14)
    plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=t, cmap=plt.cm.hot)
    plt.xlabel("$z_1$", fontsize=18)
    if subplot == 131:
        plt.ylabel("$z_2$", fontsize=18, rotation=0)
    plt.grid(True)

plt.show()
```
